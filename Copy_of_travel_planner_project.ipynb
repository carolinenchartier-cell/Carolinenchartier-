```python
from requests.models import Response
import requests
from rich import print
from rich.markdown import Markdown

def display_current_weather(location):
  """ Get the real time temperature and condition in a location"""
  api_key = "0fa35b0t8477d72d0f1c0ad4ba3fo913"
  api_url = f"https://api.shecodes.io/weather/v1/current?query={location}&key={api_key}&units=metric"

  response = requests.get(api_url)
  response_data = response.json()

  temperature= round(response_data['temperature']['current'])
  condition = response_data['condition']['description']


  print(f"\nThe current temperature in [bold]{location}[/bold] is {temperature}C, {condition}\n")
  # print(response_data) # Removed this line to avoid printing raw JSON

def generate_itinerary(origin, destination, duration, api_key, context_text):
  print(f"\n\nGenerating itinerary from {origin} to {destination} for {duration} days...\n")

  # Construct prompt with user inputs
  prompt = f"Generate a travel itinerary from {origin} to {destination} in {duration} days. This is a road trip, keep it short, less than 15 lines, add some emojis to make it more readable. Add an estimated price and weather conditions for each day."
  api_url_formatted = f"https://api.shecodes.io/ai/v1/generate?prompt={prompt}&context={context_text}&key={api_key}"

  response= requests.get(api_url_formatted)
  response_data = response.json()
  # print(response_data ['answer']) # Removed this redundant print statement
  itinerary = Markdown(response_data['answer'])
  print(itinerary)

def welcome():
  print("Welcome to the AI Travel planner")

def credit():
  print("This AI Itinerary was built by Caroline Chartier. Thank you for using it.")

welcome()

# Global variables for API setup (context and key)
context = "You are a travel specialist and know the best spots around the world."
api_key= "0fa35b0t8477d72d0f1c0ad4ba3fo913" # Replace with your valid API key

# Get user input
origin= input("What city does your trip start from?")
destination= input("What city is your trip going to?")
duration = input("How long will your trip last? (enter a number only, i.e. 5)")

if origin and destination and duration.isdigit():
  display_current_weather(origin)
  display_current_weather(destination)
  generate_itinerary(origin, destination, duration, api_key, context)
  credit()
else:
  print("Something is wrong, please try again")
```
